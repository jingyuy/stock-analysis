```python
import pandas as pd
from datetime import datetime
import pandas_datareader.data as web
import pandas_datareader as pdr
import numpy as np
import ta

path = "./"
t = pd.HDFStore(path + "tickers.hd5", "r")
tickers = t["data"]
t.close()

def save_features(ticker, features):
    fs = pd.HDFStore(path + ticker +"_features.hd5", "w")
    fs["data"] = features
    fs.close()
    
def load_features(ticker):
    fs = pd.HDFStore(path + ticker +"_features.hd5", "r")
    features = fs["data"]
    fs.close()
    return features

def load_source_data(ticker):
    fs = pd.HDFStore(path + ticker + ".hd5", "r")
    data = fs["data"]
    fs.close()
    return data

    
f = pd.DataFrame([], columns=[])
fs = pd.DataFrame([], columns=[])
for ticker in tickers['ticker']:
    data = load_source_data(ticker)
    fs = ta.add_all_ta_features(data, "open", "high", "low", "adjusted close", "volume")
    fs = fs[::-1]
    data = data[::-1]
    f["adjusted_close"] = data["adjusted close"]
    f["volume"] = data["volume"]
    f["daily_log_return"] = np.log(data["adjusted close"]) - np.log(data["adjusted close"].shift(-1))
    f["average_20_days"] = data["adjusted close"].rolling(window=20).mean().shift(-20)
    f["average_50_days"] = data["adjusted close"].rolling(window=50).mean().shift(-50)
    f["average_200_days"] = data["adjusted close"].rolling(window=200).mean().shift(-200)
    f["volume_average_20_days"] = data["volume"].rolling(window=20).mean().shift(-20)
    f["volume_average_50_days"] = data["volume"].rolling(window=50).mean().shift(-50)
    f["volume_average_200_days"] = data["volume"].rolling(window=200).mean().shift(-200)
    
    save_features(ticker, f)
    
```
